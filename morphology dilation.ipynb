## Morpholgy transformation and image segementation

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

# dilation

```python
img=cv.imread("C:/Users/admin/Downloads/j.jpg",0)
img=cv.cvtColor(img,cv.COLOR_BGR2RGB)
_,mask=cv.threshold(img,220,255,cv.THRESH_BINARY)
kernal=np.ones((5,5),np.uint8)
dilation=cv.dilate(mask,kernal,iterations=1)
erode=cv.erode(mask,kernal,iterations=1)
titles=['images','mask','dilate','erode']
images=[img,mask,dilation,erode]
for i in range(len(images)):
    plt.subplot(2,4,i+1),plt.imshow(images[i])
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
```

# Morphological opening-errosion followed by dilation

```python
open=cv.morphologyEx(img,cv.MORPH_OPEN,kernal)
```

```python
cv.imshow('img',img)
cv.imshow('morph open',open)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
titles=['images','dilate','erode','morph','open']
images=[img,dilation,erode,open]
for i in range(len(images)):
    plt.subplot(2,4,i+1),plt.imshow(images[i])
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
```

# CLOSE EROSION

```python
close=cv.morphologyEx(img,cv.MORPH_CLOSE,kernal)
```

```python
cv.imshow('img',img)
cv.imshow('morph close',close)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
titles=['images','dilate','erode','morph','close']
images=[img,dilation,erode,close]
for i in range(len(images)):
    plt.subplot(2,4,i+1),plt.imshow(images[i])
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
```

# Gradient

```python
gradient=cv.morphologyEx(img,cv.MORPH_GRADIENT,kernal)
```

```python
cv.imshow('ORIGINAL',img)
cv.imshow('MORPH GRADIENT',gradient)
cv.waitKey(0)
cv.destroyAllWindows()
```

# Tophat

```python
tophat=cv.morphologyEx(img,cv.MORPH_TOPHAT,kernal)
```

```python
cv.imshow('ORIGINAL',img)
cv.imshow('TOP HAT',tophat)
cv.waitKey(0)
cv.destroyAllWindows()
```

# Black hat

```python
Blackhat=cv.morphologyEx(img,cv.MORPH_BLACKHAT,kernal)
```

```python
cv.imshow('ORIGINAL',img)
cv.imshow('BLACK HAT',Blackhat)
cv.waitKey(0)
cv.destroyAllWindows()
```

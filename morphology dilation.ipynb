## Morpholgy transformation and image segementation

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

# dilation

```python
img=cv.imread("C:/Users/admin/Downloads/j.jpg",0)
img=cv.cvtColor(img,cv.COLOR_BGR2RGB)
_,mask=cv.threshold(img,220,255,cv.THRESH_BINARY)
kernal=np.ones((5,5),np.uint8)
dilation=cv.dilate(mask,kernal,iterations=1)
erode=cv.erode(mask,kernal,iterations=1)
titles=['images','mask','dilate','erode']
images=[img,mask,dilation,erode]
for i in range(len(images)):
    plt.subplot(2,4,i+1),plt.imshow(images[i])
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
```
